```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression

X = [[1] , [2] , [3] , [4] , [5]]
y = [0 , 0 , 1 , 1 , 1] # result

model = LogisticRegression()

model.fit(X , y)

hours = float(input("Enter how many hours you have studied = "))

result = model.predict([[hours]])[0]

if result == 1:
    print(f"If you study {hours} hours then you definately going to pass the exam")
else:
    print(f"If you study {hours} hours then you definately going to fail the exam")
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier

X = [
    [100 , 7],
    [200 , 7.5],
    [250 , 8],
    [300 , 8.5],
    [330 , 9],
    [360 , 9.5],
]

# 0 -> Apple , 1 -> Orange

y = [0,0,0,1,1,1]

model = KNeighborsClassifier(n_neighbors=3) # aas pass ke 3 neighbours ko dekhna hai

model.fit(X , y)

weight = float(input("Weight of the fruit in gm: "))
size = float(input("Size of the fruit in cm:"))

result = model.predict([[weight , size]])[0]

if result == 0:
    print(f"The fruit with weight = {weight} and size = {size} is Apple")
else:
    print(f"The fruit with weight = {weight} and size = {size} is Orange")
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier

x = [
    [7,2],
    [8,3],
    [9,8],
    [10,9]
]

y = [0 , 0 , 1 , 1]  # 0 -> Apple , 1 -> Orange

model = DecisionTreeClassifier()

model.fit(x , y)
size = float(input("Enter the size of the fruit: "))
color = float(input("Enter the color shade of the fruit (1 -> reddish , 10 -> orangish) : "))

result = model.predict([[size , color]])[0]

if(result == 0):
    print("The fruit is likely to be an Apple")
else:
    print("The fruit is likely to be an Orange")
 
```
