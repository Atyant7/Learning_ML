```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

data = pd.read_csv("student_scores.csv")

X = data[['Hours']] # Double brackets for 2D input
y = data[['Score']] # Double brackets for 2D input   # target column

model = LinearRegression()

model.fit(X, y)

predicted_score = model.predict(X)

#evaluate
mae = mean_absolute_error(y , predicted_score)
mse = mean_squared_error(y , predicted_score)
rmse = np.sqrt(mse)

#show results
print("Mean Absolute Error:",mae)
print("Mean Squared Error:" , mse)
print("Root Mean Squared Error:", rmse)

new_hour = float(input("Enter a hour: "))
new_pred = model.predict([[new_hour]])[0]
print(f"Prediction for {new_hour} is score = {new_pred}")

new_prediction = model.predict([[7]])[0]
print(f"Prediction score for {7} hour : {new_prediction}")
```
