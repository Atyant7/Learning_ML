## What is pandas
- used for Data Handling and analysis
- works best with tabular data (like excel)
- built on top on numpy

```python
import pandas as pd

data = [10, 20, 30, 40, 50]
print(data)
print()
s = pd.Series(data)
print(s)
type(s)
print()
s2 = pd.Series([10, 20, 30] , index=['a', 'b', 'c'])
print(s2)
print()

s4 = pd.Series(5, index=["x" , "y" , "z"])
print(s4)
```

Question-> represent the marks of the students in a series where index is the name of the student

```python
import pandas as pd

data = [50,60,75,89,42,98]
name = ['Ashish' , 'Avin' , 'Deva' , 'Dipen' , 'Om' , 'Raghu']
s = pd.Series(data , index=name)
print(s)
```
